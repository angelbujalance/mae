```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_json("pretrain_mae/lr_1e-3_wd_0.05_mr_0.7/loss_ncc_dict.json")

# Create figure and axis
fig, ax1 = plt.subplots()

# Plot loss
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss', color='tab:red')
ax1.plot(df["epochs"], df["loss"], color='tab:red', label='Loss')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create second y-axis for NCC
ax2 = ax1.twinx()
ax2.set_ylabel('NCC', color='tab:blue')
ax2.plot(df["epochs"], df["ncc"], color='tab:blue', label='NCC')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Title and layout
plt.title('Loss and NCC over Epochs')
fig.tight_layout()
plt.show()
```
